```python
from utils.bezier_parsec_parametrization import BezierParsecAirfoil
import numpy as np

airfoil = BezierParsecAirfoil.from_random()
print(airfoil)
from utils.get_airfoil_coordinates import get_airfoil_coordinates
from matplotlib import pyplot as plt

coordinates = get_airfoil_coordinates(airfoil)

thickness_control_points = airfoil.thickess_curves[0].control_points + airfoil.thickess_curves[1].control_points
camber_control_points = airfoil.camber_curves[0].control_points + airfoil.camber_curves[1].control_points

fig, axs = plt.subplots(3, figsize=(10, 10))

t = np.linspace(0, 1, 100)
thickness_values = [airfoil.thickness_at(t_i) for t_i in t]
camber_values = [airfoil.camber_at(t_i) for t_i in t]
axs[0].plot([thickness[0] for thickness in thickness_values], [thickness[1] for thickness in thickness_values])
axs[0].plot([coord[0] for coord in thickness_control_points], [coord[1] for coord in thickness_control_points], 'or')
axs[1].plot([camber[0] for camber in camber_values], [camber[1] for camber in camber_values])
axs[1].plot([coord[0] for coord in camber_control_points], [coord[1] for coord in camber_control_points], 'or')
axs[2].plot([coord[0] for coord in coordinates], [coord[1] for coord in coordinates])
```
